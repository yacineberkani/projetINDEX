## Ensemble de données sur les commentaires des clients de British Airways

## Aperçu:

**Cet ensemble de données contient les commentaires des clients de British Airways, extraits d' AirlineQuality via le scraping Web. L'ensemble de données est une ressource riche pour l'analyse des sentiments, permettant aux chercheurs et aux passionnés de données d'explorer et de comprendre les sentiments exprimés par les passagers de British Airways.**

## Colonnes:

- **OverallRating :** La note globale attribuée par le client.
- **ReviewHeader :** l'en-tête ou le titre de l'avis du client.
- **Name :** Le nom du client fournissant le commentaire.
- **Datetime :** la date et l’heure à laquelle le commentaire a été publié.
- **VerifiedReview :** indique si l'avis est vérifié ou non.
- **ReviewBody :** Le corps détaillé de l'avis du client.
- **TypeOfTraveller :** le type de voyageur (par exemple, affaires, loisirs).
- **SeatType :** Classe du voyageur (par exemple Affaires, Économie).
- **Route :** L'itinéraire de vol emprunté par le client.
- **DateFlown :** la date à laquelle le vol a été effectué.
- **SeatComfort :** Évaluation du confort du siège.
- **CabinStaffService :** Évaluation du service du personnel de cabine.
- **GroundService :** Évaluation du service au sol.
- **ValueForMoney:** Note pour le rapport qualité/prix.
- **Recommended :** si le client recommande British Airways.
- **Aircraft :** L'aéronef utilisé pour le vol.
- **Food&Beverages :** Évaluation des aliments et des boissons.
- **InflightEntertainment :** évaluation du divertissement en vol.
- **Wifi&Connectivity :** Note pour le wifi et la connectivité à bord.

# 2. Construction de l'index :

```python
# Choisir les colonnes à indexer (par exemple ReviewHeader et ReviewBody)
colonnes_textuelles = ["ReviewHeader", "ReviewBody"]
```

```python
import nltk
nltk.download('punkt_tab')
nltk.download('punkt')
nltk.download('stopwords')
```

```python
import re # Importe le module 're' pour les expressions régulières.
from nltk.corpus import stopwords # Importe 'stopwords' de nltk.corpus pour supprimer les mots vides.
from nltk.tokenize import word_tokenize # Importe 'word_tokenize' pour la tokenisation du texte.

def preprocess_document(document_text): # Définit une fonction pour prétraiter un document.
    # Convertir en minuscules
    text = document_text.lower() # Convertit le texte en minuscules.
    # Supprimer la ponctuation
    text = re.sub(r'[^\w\s]', '', text) # Supprime la ponctuation du texte en utilisant une expression régulière.
    # Tokenization
    tokens = word_tokenize(text) # Divise le texte en tokens (mots individuels).
    # Supprimer les mots vides
    stop_words = set(stopwords.words('english')) # Crée un ensemble de mots vides anglais.
    filtered_tokens = [word for word in tokens if word not in stop_words] # Supprime les mots vides des tokens.
    return filtered_tokens # Retourne les tokens filtrés.
```

```python
# Construire le dictionnaire
inverted_index = {}
```

```python
def construire_index_inverse(df, colonnes_textuelles):
    # Dictionnaire qui va contenir l'index inversé
    inverted_index = {}

    for doc_id, row in df.iterrows():
        # Concaténer le texte des deux colonnes
        texte_document = ""
        for col in colonnes_textuelles:
            # Éviter l'erreur si la cellule est NaN
            valeur_colonne = row[col] if pd.notnull(row[col]) else ""
            texte_document += str(valeur_colonne) + " "

        # Nettoyer et tokenizer
        tokens = preprocess_document(texte_document)

        # Parcourir chaque token et enregistrer sa position
        for pos, token in enumerate(tokens):
            # Initialiser la clé du token si elle n'existe pas
            if token not in inverted_index:
                inverted_index[token] = {}

            # Ajouter doc_id dans la liste s'il n'y est pas
            if doc_id not in inverted_index[token]:
                inverted_index[token][doc_id] = []

            # Ajouter la position
            inverted_index[token][doc_id].append(pos)

    return inverted_index
```

```python
# Construire l'index inversé
inverted_index = construire_index_inverse(df, colonnes_textuelles)

# Vérification : afficher les 5 premiers mots de l'index inversé
for i, (mot, occurrences) in enumerate(inverted_index.items()):
    print(mot, ":", occurrences)
    if i >= 4:
        break
```

# 3. Implémentation de la recherche :


```python
# Recherche d’un mot unique
def search_single_word(inverted_index, word):
    """
    Recherche un mot unique dans l'index inversé.
    :param inverted_index: dict (structure de l'index inversé)
    :param word: str (le mot à rechercher)
    :return: dict {doc_id: [positions], ...}
    """
    # Nettoyer le mot avant de le chercher, par exemple passer en minuscule
    word = word.lower()

    # Vérifier si le mot existe dans l'index
    if word in inverted_index:
        return inverted_index[word]
    else:
        return {}
```

```python
# Utilisation
mot_a_chercher = "service"
resultats = search_single_word(inverted_index, mot_a_chercher)

for doc_id, positions in resultats.items():
    print(f"Document {doc_id}, positions : {positions}")
```

```python
# Recherche de plusieurs mots (intersection, union, etc.)
#Intersection de mots
def search_multiple_words_and(inverted_index, words_list):
    """
    Recherche l'intersection des documents contenant tous les mots.
    :param inverted_index: dict (index inversé)
    :param words_list: liste de mots à rechercher
    :return: dict {doc_id: {mot: [positions], ...}, ...}
             Pour chaque doc_id, on indique, pour chaque mot, les positions où il apparaît
    """
    # Normaliser tous les mots (minuscule, etc.)
    words_list = [w.lower() for w in words_list]

    # Récupérer les résultats individuels pour chaque mot
    list_of_results = []
    for w in words_list:
        if w in inverted_index:
            list_of_results.append(inverted_index[w])
        else:
            # Si un mot n'existe pas du tout, l'intersection est vide
            return {}

    # Maintenant, il faut faire l'intersection sur les doc_id
    common_doc_ids = set(list_of_results[0].keys())
    for dico_doc_positions in list_of_results[1:]:
        common_doc_ids = common_doc_ids.intersection(dico_doc_positions.keys())

    # Construire la structure de retour
    final_results = {}
    for doc_id in common_doc_ids:
        final_results[doc_id] = {}
        # Pour chaque mot, on stocke les positions
        for w in words_list:
            final_results[doc_id][w] = inverted_index[w][doc_id]

    return final_results
```

```python
# Utilisation
mots_a_chercher = ["service", "level"]  # Définit une liste de mots à rechercher : "service" et "level"
resultats = search_multiple_words_and(inverted_index, mots_a_chercher)  # Appelle la fonction search_multiple_words_and pour trouver les documents contenant tous les mots de la liste et stocke le résultat dans 'resultats'

print("\n=== Résultats de la recherche ===")  # Affiche un titre pour les résultats de la recherche
if not resultats:  # Vérifie si le dictionnaire 'resultats' est vide (aucun document trouvé)
    print("Aucun document trouvé")  # Affiche un message indiquant qu'aucun document n'a été trouvé
else:  # Si 'resultats' n'est pas vide (documents trouvés)
    # Itère sur chaque document (doc_id) et les informations associées (infos) dans 'resultats'
    for doc_id, infos in resultats.items():
        print(f"\nDocument {doc_id} contient tous les mots : {mots_a_chercher}")  # Affiche le numéro du document et la liste des mots recherchés qu'il contient
        # Itère sur chaque mot (mot) et ses positions (positions) dans le document courant
        for mot, positions in infos.items():
            print(f"   -> '{mot}' aux positions {positions}")  # Affiche le mot et les positions où il apparaît dans le document
```

```python
#  Union de mots (OR)
def search_multiple_words_or(inverted_index, words_list):
    """
    Recherche la liste des documents contenant au moins un des mots.
    :return: dict {doc_id: {mot: [positions], ...}, ...}
    """
    words_list = [w.lower() for w in words_list]
    final_results = {}

    for w in words_list:
        if w in inverted_index:
            for doc_id, positions in inverted_index[w].items():
                # Initialiser
                if doc_id not in final_results:
                    final_results[doc_id] = {}
                # Ajouter le mot et ses positions
                final_results[doc_id][w] = positions

    return final_results
```

```python
# Utilisation
mots_a_chercher = ["service", "friendly"]  # Définit une liste de mots à rechercher : "service" et "friendly"
resultats = search_multiple_words_or(inverted_index, mots_a_chercher)  # Appelle la fonction search_multiple_words_or pour trouver les documents contenant tous les mots de la liste et stocke le résultat dans 'resultats'

print("\n=== Résultats de la recherche ===")  # Affiche un titre pour les résultats de la recherche
if not resultats:  # Vérifie si le dictionnaire 'resultats' est vide (aucun document trouvé)
    print("Aucun document trouvé")  # Affiche un message indiquant qu'aucun document n'a été trouvé
else:  # Si 'resultats' n'est pas vide (documents trouvés)
    # Itère sur les 100 premières documents (doc_id)
    for doc_id, infos in list(resultats.items())[:100]:
        print(f"\nDocument {doc_id}")  # Affiche le numéro du document
        # Itère sur chaque mot (mot) et ses positions (positions) dans le document courant
        for mot, positions in infos.items():
            print(f"   -> '{mot}' aux positions {positions}")  # Affiche le mot et les positions où il apparaît dans le document
```

```python
# Recherche d’une expression exacte (Phrase Search)
def search_phrase(inverted_index, phrase):
    """
    Recherche d'une expression exacte en s'appuyant sur les positions.
    ex: "service was friendly"
    """
    # Applique la fonction de prétraitement 'preprocess_document' à la phrase et stocke le résultat (liste de mots nettoyés) dans la variable 'words'.
    words = preprocess_document(phrase)

    if not words:
        return {}

    # Récupérer la liste des doc_id pour chaque mot
    posting_lists = []
    for w in words:
        if w in inverted_index:
            posting_lists.append(inverted_index[w])
        else:
            # Si un mot de la phrase n'existe pas, la recherche est vide
            return {}

    # Étape 1 : intersection des doc_id
    common_doc_ids = set(posting_lists[0].keys())
    for pl in posting_lists[1:]:
        common_doc_ids = common_doc_ids.intersection(pl.keys())

    # Étape 2 : vérifier la séquence des positions dans chaque doc_id
    results = {}
    for doc_id in common_doc_ids:
        # On récupère toutes les positions du 1er mot
        positions_word1 = posting_lists[0][doc_id]
        # On va tester pour chaque position du 1er mot si les mots suivants sont en positions +1, +2, etc.

        for pos1 in positions_word1:
            match_found = True
            # On vérifie si, pour chaque mot suivant, la position est pos1 + i
            for i in range(1, len(words)):
                # doc_id est le même, on check dans posting_lists[i] si pos1 + i est présent
                if (pos1 + i) not in posting_lists[i][doc_id]:
                    match_found = False
                    break

            if match_found:
                # On a trouvé un match à partir de pos1
                if doc_id not in results:
                    results[doc_id] = []
                results[doc_id].append(pos1)

    return results
```

```python
# Utilisation
expression = "service was friendly"  # Définit l'expression exacte à rechercher : "service was jsk friendly"
resultats = search_phrase(inverted_index, expression)  # Appelle la fonction search_phrase pour trouver les documents contenant l'expression exacte et stocke le résultat dans 'resultats'

print("\n=== Résultats de la recherche ===")  # Affiche un titre pour les résultats de la recherche
if not resultats:  # Vérifie si le dictionnaire 'resultats' est vide (aucun document trouvé)
    print("Aucun document trouvé")  # Affiche un message indiquant qu'aucun document n'a été trouvé
else:  # Si 'resultats' n'est pas vide (documents trouvés)
    # Itère sur chaque document (doc_id) et les positions de départ de l'expression (start_positions) dans 'phrase_results'
    for doc_id, start_positions in phrase_results.items():  # **Attention: phrase_results devrait être resultats**
        print(f"\nDocument {doc_id} contient l'expression '{expression}' aux positions de départ : {start_positions}")  # Affiche le numéro du document et les positions où l'expression commence
```

```python
def get_excerpt(df, doc_id, start_pos, window=5):
    """
    Récupère 'window' mots avant et après la position 'start_pos'
    pour donner un petit extrait autour de cette position.
    """
    # Concaténer le texte du doc
    row = df.iloc[doc_id]
    texte_document = (str(row["ReviewHeader"]) + " " + str(row["ReviewBody"])).lower()
    tokens = texte_document.split()

    # Extraire la fenêtre
    left_index = max(start_pos - window, 0)
    right_index = min(start_pos + window + 1, len(tokens))
    excerpt = tokens[left_index:right_index]

    return " ".join(excerpt)
```
